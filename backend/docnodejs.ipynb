Bước 3 : truyền vào một hàm callback 
Bươc 4 : trong hàm callback sẽ trả về hai tham số là req ,res
    Req se chua toan bo thong tin client gui len  method la gif get pot put path 
    Query parameter la gi  duong dan la gi , payload la gi  , body la cai gi
    Noi chung la tat ca thu gi tu client  gui len no se nam trong object nay 
    Res khi bạn gọi hàm writeHead sẽ đính vào header hiển thị status 200 
    Sau đó định dạng chuối trả về client điểm cuôi là một json
    Res sẽ quyết định trả cái gì về cho client  

Khi request gửi lên nó sẽ chờ ở server khi server quyết định kết thúc thì nó mới kết thúc cái request
 Res.End có nghĩa là kết thúc và trả về dữ liệu
 




```python
const http = require('node:http');

// Create a local server to receive data from
const server = http.createServer();

// Listen to the request event
server.on('request', (request, res) => {
  res.writeHead(200, { 'Content-Type': 'application/json' });
  res.end(JSON.stringify({
    data: 'Hello World!',
  }));
});

server.listen(8000);
```
